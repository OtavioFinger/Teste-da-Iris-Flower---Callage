# Gráfico Inicial a partir dos dados

```python
plt.figure(figsize=(6,4))

plt.scatter(df['petal_length'],df['petal_width'], color='purple', alpha=0.7)

#alpha é a transparência do seu gráfico

font1 = {'family':'serif','color':'black','size':14}
font2 = {'family':'serif','color':'black','size':14}

plt.xlabel('Comprimento da Pétala(cm)', fontdict = font1)
plt.ylabel('Largura da Pétala(cm)', fontdict = font2)
plt.title('Gráfico de Pontos - Dataset Iris')


plt.grid()
plt.show()
```

# Gráfico de *Dispersão*

```python
cores_map = {
    'Iris-setosa' : 'purple',
    'Iris-versicolor' : 'red',
    'Iris-virginica' : 'blue',
}

df['cor_ponto'] = df['species'].map(cores_map)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['petal_length'],df['petal_width'], c=df['cor_ponto'], alpha = 0.7)

plt.xlabel('Comprimento da petala(cm):')
plt.ylabel('Largura da petala(cm):')
plt.title("Visualização Cartesiana de Espécies de Iris")

plt.grid()
plt.show()
```
